## Helper Function 1: Get All Customer Data

Retrieve all relevant data for one customer in a single pass.
Avoids redundant lookups later.

```python
# Helper Function - Get Customer Data (OPTIMIZED)

# Build lookup dictionaries ONCE after data load (highly efficient)
def build_lookup_maps():
    """
    Pre-compute all lookup dictionaries to avoid repeated DataFrame filtering.
    Called once after data load, before any customer evaluation.
    """
    # Store full rows as dict by customer_id
    customer_map = {row['customer_id']: row for _, row in customers_df.iterrows()}
    
    # Group by customer_id to create O(1) lookups
    id_map = {cid: grp for cid, grp in id_verifications_df.groupby('customer_id')}
    doc_map = {cid: grp for cid, grp in documents_df.groupby('customer_id')}
    screening_map = {cid: grp for cid, grp in screenings_df.groupby('customer_id')}
    txn_map = {cid: grp for cid, grp in transactions_df.groupby('customer_id')}
    
    # Create UBO lookup from JSON
    ubo_map = {}
    for record in ubo_data:
        cid = record.get('customer_id')
        if cid:
            ubo_map[cid] = record.get('ubos', [])
    
    # Create empty template DataFrames for missing data
    empty_ids = id_verifications_df.iloc[0:0]
    empty_docs = documents_df.iloc[0:0]
    empty_screenings = screenings_df.iloc[0:0]
    empty_txns = transactions_df.iloc[0:0]
    
    return {
        'customer_map': customer_map,
        'id_map': id_map,
        'doc_map': doc_map,
        'screening_map': screening_map,
        'txn_map': txn_map,
        'ubo_map': ubo_map,
        'empty_ids': empty_ids,
        'empty_docs': empty_docs,
        'empty_screenings': empty_screenings,
        'empty_txns': empty_txns
    }

# Build maps ONCE (this is expensive, but done only once)
print("Building lookup maps... (one-time cost)")
lookup_maps = build_lookup_maps()
print("✓ Lookup maps built - ready for O(1) customer lookups")

def get_customer_data(customer_id):
    """
    Retrieve all relevant data for a customer using O(1) lookups.
    Replaces repeated DataFrame filtering with pre-computed maps.
    
    Returns dict with all customer info, or None if not found.
    """
    
    # O(1) lookup in pre-built customer map
    customer_row = lookup_maps['customer_map'].get(customer_id)
    if customer_row is None:
        return None
    
    # O(1) lookups for all related data, with empty defaults
    ids = lookup_maps['id_map'].get(customer_id, lookup_maps['empty_ids'])
    docs = lookup_maps['doc_map'].get(customer_id, lookup_maps['empty_docs'])
    screenings = lookup_maps['screening_map'].get(customer_id, lookup_maps['empty_screenings'])
    transactions = lookup_maps['txn_map'].get(customer_id, lookup_maps['empty_txns'])
    ubos = [{'customer_id': customer_id, 'ubos': lookup_maps['ubo_map'][customer_id]}] \
           if customer_id in lookup_maps['ubo_map'] else []
    
    # Return all customer data as dictionary
    return {
        # Customer master
        'customer_id': customer_row['customer_id'],
        'customer_name': customer_row['customer_name'],
        'entity_type': customer_row['entity_type'],
        'jurisdiction': customer_row['jurisdiction'],
        'risk_rating': customer_row['risk_rating'],
        'account_open_date': customer_row['account_open_date'],
        'last_kyc_review_date': customer_row['last_kyc_review_date'],
        
        # Related data (as dataframes/lists)
        'ids': ids,
        'documents': docs,
        'screenings': screenings,
        'transactions': transactions,
        'ubos': ubos
    }

print("✓ get_customer_data helper function defined (optimized)")
```

# Dimension 1: Identity Verification

**Purpose:** Verify customer has valid, current proof of identity.

**Checks:**
1. ID exists on file
2. ID not expired
3. ID verification recent (age check)
4. ID type valid for jurisdiction

**Decision Logic:**
- Missing ID → ESCALATE (separate reason code: MISSING_ID)
- Expired ID → ESCALATE (reason code: EXPIRED_ID)
- Verification too old → ESCALATE (reason code: VERIFICATION_TOO_OLD)
- Invalid ID type → ESCALATE (reason code: INVALID_ID_TYPE)
- Any other issue → ESCALATE

**Thresholds:**
- ID verification age:
  - High-Risk: <180 days (6 months)
  - Medium-Risk: <270 days (9 months)
  - Low-Risk: <365 days (12 months)
- ID expiry:
  - Standard: Must expire >1825 days from now (5 years)
  - UK: Must have >180 days remaining (stricter)
- Valid ID types by jurisdiction:
  - UK/EU: Passport, National ID
  - US: Passport, Driver License, State ID
  - Offshore: Passport only

## Function: check_id_verification()

Evaluate identity verification for one customer.

**Inputs:**
- customer_data: dictionary returned by get_customer_data()

**Returns:**
Dictionary with:
- status: PASS | ESCALATE
- reason_code: MISSING_ID | EXPIRED_ID | VERIFICATION_TOO_OLD | INVALID_ID_TYPE | PASS
- reason: Human-readable explanation
- days_until_expiry: Days remaining (negative if expired)
- days_since_verification: Days since ID was verified
- id_type: Type of ID (Passport, Driver License, etc.)
- verification_date: When ID was verified
- expiry_date: When ID expires

```python
# Dimension 1 - Identity Verification Function

def check_id_verification(customer_data):
    """
    Evaluate identity verification for a customer.
    
    Returns dict with:
    - status: PASS | ESCALATE
    - reason_code: specific code (MISSING_ID, EXPIRED_ID, etc.)
    - reason: human-readable explanation
    - supporting data (dates, ID type, days remaining)
    """
    
    customer_id = customer_data['customer_id']
    jurisdiction = customer_data['jurisdiction']
    risk_rating = customer_data['risk_rating']
    ids = customer_data['ids']
    
    # Check 1: ID exists
    if len(ids) == 0:
        return {
            "status": "ESCALATE",
            "reason_code": "MISSING_ID",
            "reason": "No ID on file",
            "days_until_expiry": None,
            "days_since_verification": None,
            "id_type": None,
            "verification_date": None,
            "expiry_date": None
        }
    
    # Get most recent ID (by verification date)
    latest_id = ids.loc[ids['verification_date'].idxmax()]
    
    expiry_date = latest_id['expiry_date']
    verification_date = latest_id['verification_date']
    id_type = latest_id['id_type']
    
    # Check 2: ID not expired
    days_until_expiry = (expiry_date - EVALUATION_DATE).days
    
    if days_until_expiry < 0:
        return {
            "status": "ESCALATE",
            "reason_code": "EXPIRED_ID",
            "reason": "ID EXPIRED {0} days ago".format(abs(days_until_expiry)),
            "days_until_expiry": days_until_expiry,
            "days_since_verification": (EVALUATION_DATE - verification_date).days,
            "id_type": id_type,
            "verification_date": verification_date,
            "expiry_date": expiry_date
        }
    
    # Check 3: ID expiry approaching (jurisdiction-specific)
    if jurisdiction == "UK":
        expiry_threshold = 180
    else:
        expiry_threshold = 365 * 5
    
    if days_until_expiry < expiry_threshold:
        return {
            "status": "ESCALATE",
            "reason_code": "EXPIRED_ID" if jurisdiction == "UK" else "EXPIRY_APPROACHING",
            "reason": "ID expires in {0} days (threshold: {1})".format(days_until_expiry, expiry_threshold),
            "days_until_expiry": days_until_expiry,
            "days_since_verification": (EVALUATION_DATE - verification_date).days,
            "id_type": id_type,
            "verification_date": verification_date,
            "expiry_date": expiry_date
        }
    
    # Check 4: Verification not too old (risk-based)
    days_since_verification = (EVALUATION_DATE - verification_date).days
    
    if risk_rating == "High":
        verification_threshold = 180
    elif risk_rating == "Medium":
        verification_threshold = 270
    else:  # Low
        verification_threshold = 365
    
    if days_since_verification > verification_threshold:
        return {
            "status": "ESCALATE",
            "reason_code": "VERIFICATION_TOO_OLD",
            "reason": "ID verification is {0} days old (threshold: {1} for {2}-risk)".format(
                days_since_verification, verification_threshold, risk_rating),
            "days_until_expiry": days_until_expiry,
            "days_since_verification": days_since_verification,
            "id_type": id_type,
            "verification_date": verification_date,
            "expiry_date": expiry_date
        }
    
    # Check 5: ID type valid for jurisdiction
    valid_types = {
        "UK": ["Passport", "National ID"],
        "EU": ["Passport", "National ID"],
        "US": ["Passport", "Driver License", "State ID"],
        "Offshore": ["Passport"]
    }
    
    acceptable_types = valid_types.get(jurisdiction, ["Passport", "National ID"])
    
    if id_type not in acceptable_types:
        return {
            "status": "ESCALATE",
            "reason_code": "INVALID_ID_TYPE",
            "reason": "{0} not acceptable for {1} (valid types: {2})".format(
                id_type, jurisdiction, ", ".join(acceptable_types)),
            "days_until_expiry": days_until_expiry,
            "days_since_verification": days_since_verification,
            "id_type": id_type,
            "verification_date": verification_date,
            "expiry_date": expiry_date
        }
    
    # All checks passed
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "ID current and valid: {0} expires in {1} days, verified {2} days ago".format(
            id_type, days_until_expiry, days_since_verification),
        "days_until_expiry": days_until_expiry,
        "days_since_verification": days_since_verification,
        "id_type": id_type,
        "verification_date": verification_date,
        "expiry_date": expiry_date
    }

print("✓ check_id_verification function defined")
```

# Dimension 2: Proof of Address

**Purpose:** Verify customer has current proof of address.

**Checks:**
1. PoA document exists
2. PoA not expired
3. PoA recent (<24 months old)

**Decision Logic:**
- Missing PoA → ESCALATE
- Expired PoA → ESCALATE
- PoA >24 months old → REMEDIATE
- PoA current (<24 months) → PASS

**Thresholds:**
- All jurisdictions: <24 months old (730 days)

```python
# Dimension 2: Proof of Address

def check_proof_of_address(customer_data):
    """
    Evaluate proof of address for a customer.
    
    Returns dict with status, reason_code, and supporting data.
    """
    
    documents = customer_data['documents']
    
    # Filter to PoA documents only
    poa_docs = documents[documents['document_type'] == 'Proof of Address']
    
    # Check 1: PoA exists
    if len(poa_docs) == 0:
        return {
            "status": "ESCALATE",
            "reason_code": "MISSING_POA",
            "reason": "No Proof of Address on file",
            "days_since_upload": None,
            "upload_date": None,
            "expiry_date": None
        }
    
    # Get most recent PoA
    latest_poa = poa_docs.loc[poa_docs['upload_date'].idxmax()]
    
    upload_date = latest_poa['upload_date']
    expiry_date = latest_poa['expiry_date']
    
    # Check 2: PoA not expired
    days_until_expiry = (expiry_date - EVALUATION_DATE).days
    
    if days_until_expiry < 0:
        return {
            "status": "ESCALATE",
            "reason_code": "POA_EXPIRED",
            "reason": "Proof of Address expired {0} days ago".format(abs(days_until_expiry)),
            "days_since_upload": (EVALUATION_DATE - upload_date).days,
            "upload_date": upload_date,
            "expiry_date": expiry_date
        }
    
    # Check 3: PoA recent enough (24 months = 730 days)
    days_since_upload = (EVALUATION_DATE - upload_date).days
    poa_threshold = 730  # 24 months
    
    if days_since_upload > poa_threshold:
        return {
            "status": "REMEDIATE",
            "reason_code": "POA_AGING",
            "reason": "Proof of Address is {0} days old (threshold: {1})".format(days_since_upload, poa_threshold),
            "days_since_upload": days_since_upload,
            "upload_date": upload_date,
            "expiry_date": expiry_date
        }
    
    # All checks passed
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "Proof of Address current: {0} days old".format(days_since_upload),
        "days_since_upload": days_since_upload,
        "upload_date": upload_date,
        "expiry_date": expiry_date
    }

print("✓ check_proof_of_address function defined")
```

# Dimension 3: AML/Sanctions Screening

**Purpose:** Verify customer has current AML/sanctions screening.

**Checks:**
1. Screening exists
2. Screening result (no sanctions hits)
3. Screening recent (age-based by risk tier)

**Decision Logic:**
- Missing screening → ESCALATE
- Hit on sanctions list → ESCALATE (SCREENING_HIT)
- Overdue screening → ESCALATE (SCREENING_OVERDUE)
- Current screening → PASS

**Thresholds (per RC-001):**
- High-Risk: <90 days old
- Medium-Risk: <180 days old
- Low-Risk: <365 days old (12 months)

```python
# Dimension 3: AML/Sanctions Screening

def check_screening(customer_data):
    """
    Evaluate AML/sanctions screening for a customer.
    
    Returns dict with status, reason_code, and supporting data.
    """
    
    screenings = customer_data['screenings']
    risk_rating = customer_data['risk_rating']
    
    # Check 1: Screening exists
    if len(screenings) == 0:
        return {
            "status": "ESCALATE",
            "reason_code": "MISSING_SCREENING",
            "reason": "No AML/sanctions screening on file",
            "screening_date": None,
            "screening_result": None,
            "days_since_screening": None,
            "list_reference": None
        }
    
    # Get most recent screening
    latest_screening = screenings.loc[screenings['screening_date'].idxmax()]
    
    screening_date = latest_screening['screening_date']
    screening_result = latest_screening['screening_result']
    list_reference = latest_screening['list_reference']
    
    # Check 2: No sanctions hits
    if screening_result == 'Hit':
        return {
            "status": "ESCALATE",
            "reason_code": "SCREENING_HIT",
            "reason": "SANCTIONS HIT on {0}".format(list_reference),
            "screening_date": screening_date,
            "screening_result": screening_result,
            "days_since_screening": (EVALUATION_DATE - screening_date).days,
            "list_reference": list_reference
        }
    
    # Check 3: Screening recent (risk-based threshold)
    days_since_screening = (EVALUATION_DATE - screening_date).days
    
    if risk_rating == "High":
        screening_threshold = 90  # 90 days per RC-001
    elif risk_rating == "Medium":
        screening_threshold = 180
    else:  # Low
        screening_threshold = 365
    
    if days_since_screening > screening_threshold:
        return {
            "status": "ESCALATE",
            "reason_code": "SCREENING_OVERDUE",
            "reason": "Screening is {0} days old (threshold: {1} for {2}-risk)".format(
                days_since_screening, screening_threshold, risk_rating),
            "screening_date": screening_date,
            "screening_result": screening_result,
            "days_since_screening": days_since_screening,
            "list_reference": list_reference
        }
    
    # All checks passed
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "Screening current: {0} ({1} days old)".format(list_reference, days_since_screening),
        "screening_date": screening_date,
        "screening_result": screening_result,
        "days_since_screening": days_since_screening,
        "list_reference": list_reference
    }

print("✓ check_screening function defined")
```

# Dimension 4: Beneficial Ownership (UBO)

**Purpose:** Verify beneficial owners identified and verified (for legal entities).

**Checks:**
1. Applies only to LEGAL_ENTITY customers (N/A for individuals)
2. UBO data exists
3. All UBOs verified within 24 months
4. Offshore special rule: flag for registry corroboration

**Decision Logic (for LEGAL_ENTITY only):**
- Individual → N/A
- Missing UBO → ESCALATE
- UBO stale (>24 months) → ESCALATE
- Offshore + current UBO → REMEDIATE (needs registry docs)
- Current UBO → PASS

**Thresholds (per RC-003):**
- All jurisdictions: <24 months (730 days)
- Offshore: <24 months + requires corporate registry corroboration

```python
# Dimension 4: Beneficial Ownership (UBO)

def check_beneficial_ownership(customer_data):
    """
    Evaluate beneficial ownership verification for a customer.
    
    Returns dict with status, reason_code, and supporting data.
    """
    
    entity_type = customer_data['entity_type']
    jurisdiction = customer_data['jurisdiction']
    ubos = customer_data['ubos']
    
    # Check 1: Applies only to legal entities
    if entity_type != 'LEGAL_ENTITY':
        return {
            "status": "N/A",
            "reason_code": "N/A",
            "reason": "Individual customer - UBO check not applicable",
            "ubo_count": 0,
            "most_recent_verification": None,
            "oldest_verification": None
        }
    
    # Check 2: UBO exists
    if len(ubos) == 0:
        return {
            "status": "ESCALATE",
            "reason_code": "MISSING_UBO",
            "reason": "No beneficial owner data on file",
            "ubo_count": 0,
            "most_recent_verification": None,
            "oldest_verification": None
        }
    
    # Get UBO data from JSON
    ubo_list = ubos[0]['ubos'] if len(ubos) > 0 else []
    
    # Check 3: All UBOs verified within 24 months
    ubo_threshold = 730  # 24 months
    verification_dates = [datetime.strptime(u['verification_date'], '%Y-%m-%d') for u in ubo_list]
    
    if len(verification_dates) == 0:
        return {
            "status": "ESCALATE",
            "reason_code": "MISSING_UBO",
            "reason": "No UBO verification dates found",
            "ubo_count": 0,
            "most_recent_verification": None,
            "oldest_verification": None
        }
    
    most_recent = max(verification_dates)
    oldest = min(verification_dates)
    
    days_since_oldest = (EVALUATION_DATE - oldest).days
    
    if days_since_oldest > ubo_threshold:
        return {
            "status": "ESCALATE",
            "reason_code": "UBO_STALE",
            "reason": "At least one UBO not verified for {0} days (threshold: {1})".format(
                days_since_oldest, ubo_threshold),
            "ubo_count": len(ubo_list),
            "most_recent_verification": most_recent,
            "oldest_verification": oldest
        }
    
    # Check 4: Offshore special rule
    if jurisdiction == "Offshore":
        return {
            "status": "REMEDIATE",
            "reason_code": "OFFSHORE_UBO",
            "reason": "Offshore entity: UBOs current but requires corporate registry corroboration",
            "ubo_count": len(ubo_list),
            "most_recent_verification": most_recent,
            "oldest_verification": oldest
        }
    
    # All checks passed
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "UBO verified: {0} beneficial owner(s), most recent {1} days ago".format(
            len(ubo_list), (EVALUATION_DATE - most_recent).days),
        "ubo_count": len(ubo_list),
        "most_recent_verification": most_recent,
        "oldest_verification": oldest
    }

print("✓ check_beneficial_ownership function defined")
```

# Dimension 5: Account Activity

**Purpose:** Verify customer account is active (not dormant).

**Checks:**
1. Activity status = Active
2. Last transaction within 24 months

**Decision Logic:**
- Active <24 months → PASS
- At-Risk 18-24 months → REMEDIATE
- Dormant >24 months → FLAG (context dimension, not failure)

**Note:** This is a FLAG (informational), not ESCALATE (compliance issue).

```python
# Dimension 5: Account Activity

def check_account_activity(customer_data):
    """
    Evaluate account activity status for a customer.
    
    Returns dict with status and supporting data.
    """
    
    transactions = customer_data['transactions']
    
    # Check activity
    if len(transactions) == 0:
        return {
            "status": "FLAG",
            "reason_code": "NO_TRANSACTION_DATA",
            "reason": "No transaction data on file",
            "last_txn_date": None,
            "days_since_txn": None,
            "activity_status": None
        }
    
    last_txn = transactions.iloc[0]
    last_txn_date = last_txn['last_txn_date']
    activity_status = last_txn['activity_status']
    
    days_since_txn = (EVALUATION_DATE - last_txn_date).days
    
    # Evaluate based on activity status and recency
    if activity_status == 'Active' and days_since_txn < 730:  # 24 months
        return {
            "status": "PASS",
            "reason_code": "PASS",
            "reason": "Account active: last transaction {0} days ago".format(days_since_txn),
            "last_txn_date": last_txn_date,
            "days_since_txn": days_since_txn,
            "activity_status": activity_status
        }
    
    if days_since_txn >= 548 and days_since_txn < 730:  # 18-24 months
        return {
            "status": "REMEDIATE",
            "reason_code": "ACTIVITY_AT_RISK",
            "reason": "Account at risk of dormancy: {0} days since last transaction".format(days_since_txn),
            "last_txn_date": last_txn_date,
            "days_since_txn": days_since_txn,
            "activity_status": activity_status
        }
    
    if days_since_txn >= 730:  # >24 months
        return {
            "status": "FLAG",
            "reason_code": "ACCOUNT_DORMANT",
            "reason": "Account dormant: {0} days since last transaction".format(days_since_txn),
            "last_txn_date": last_txn_date,
            "days_since_txn": days_since_txn,
            "activity_status": activity_status
        }
    
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "Account active".format(days_since_txn),
        "last_txn_date": last_txn_date,
        "days_since_txn": days_since_txn,
        "activity_status": activity_status
    }

print("✓ check_account_activity function defined")
```

# Dimension 6: Prior KYC Review

**Purpose:** Verify customer has been KYC reviewed and review is current.

**Checks:**
1. KYC review exists (not NULL)
2. KYC review recent (varies by risk tier)
3. Approaching refresh threshold

**Decision Logic:**
- Never reviewed → ESCALATE
- Overdue (>threshold) → ESCALATE
- Approaching (80% of threshold) → REMEDIATE
- Current → PASS

**Thresholds (vary by risk):**
- High-Risk: <365 days (1 year)
- Medium-Risk: <730 days (2 years)
- Low-Risk: <1095 days (3 years)

```python
# Dimension 6: Prior KYC Review

def check_kyc_review(customer_data):
    """
    Evaluate prior KYC review status for a customer.
    
    Returns dict with status, reason_code, and supporting data.
    """
    
    risk_rating = customer_data['risk_rating']
    last_kyc_review = customer_data['last_kyc_review_date']
    
    # Check 1: KYC review exists
    if pd.isna(last_kyc_review):
        return {
            "status": "ESCALATE",
            "reason_code": "NO_KYC_REVIEW",
            "reason": "No prior KYC review on file",
            "last_review_date": None,
            "days_since_review": None,
            "threshold_days": None
        }
    
    # Determine threshold based on risk rating
    if risk_rating == "High":
        threshold_days = 365
    elif risk_rating == "Medium":
        threshold_days = 730
    else:  # Low
        threshold_days = 1095
    
    days_since_review = (EVALUATION_DATE - last_kyc_review).days
    
    # Check 2: KYC review overdue
    if days_since_review > threshold_days:
        return {
            "status": "ESCALATE",
            "reason_code": "KYC_OVERDUE",
            "reason": "KYC review overdue: {0} days old (threshold: {1} for {2}-risk)".format(
                days_since_review, threshold_days, risk_rating),
            "last_review_date": last_kyc_review,
            "days_since_review": days_since_review,
            "threshold_days": threshold_days
        }
    
    # Check 3: KYC review approaching threshold (80%)
    approaching_threshold = threshold_days * 0.8
    
    if days_since_review > approaching_threshold:
        return {
            "status": "REMEDIATE",
            "reason_code": "KYC_APPROACHING",
            "reason": "KYC review approaching renewal: {0} days old (threshold: {1})".format(
                days_since_review, threshold_days),
            "last_review_date": last_kyc_review,
            "days_since_review": days_since_review,
            "threshold_days": threshold_days
        }
    
    # All checks passed
    return {
        "status": "PASS",
        "reason_code": "PASS",
        "reason": "KYC review current: {0} days old (threshold: {1})".format(
            days_since_review, threshold_days),
        "last_review_date": last_kyc_review,
        "days_since_review": days_since_review,
        "threshold_days": threshold_days
    }

print("✓ check_kyc_review function defined")
```

# Dimension 7: Data Quality & Sanity Checks

**Purpose:** Validate data integrity and internal consistency.

**Checks:**
1. Required fields not NULL (risk_rating, jurisdiction)
2. Values valid (risk_rating in [High/Medium/Low], jurisdiction valid)
3. Dates logical (verification ≤ expiry, no future dates, no negative days)
4. No contradictions (e.g., account opened in future)

**Decision Logic:**
- Any sanity check fails → FLAG
- All pass → PASS

**Note:** This is separate from KYC status. A customer can ESCALATE on KYC but PASS on data quality.

```python
# Dimension 7: Data Quality & Sanity Checks

def check_data_quality(customer_data):
    """
    Validate data integrity and internal consistency.
    
    Returns dict with status and list of issues found.
    """
    
    issues = []
    
    # Check 1: Required fields exist
    if pd.isna(customer_data['risk_rating']):
        issues.append("Missing risk_rating")
    
    if pd.isna(customer_data['jurisdiction']):
        issues.append("Missing jurisdiction")
    
    # Check 2: Valid values
    valid_risk = ["High", "Medium", "Low"]
    if customer_data['risk_rating'] not in valid_risk:
        issues.append("Invalid risk_rating: {0}".format(customer_data['risk_rating']))
    
    valid_jurisdictions = ["US", "EU", "UK", "Offshore"]
    if customer_data['jurisdiction'] not in valid_jurisdictions:
        issues.append("Invalid jurisdiction: {0}".format(customer_data['jurisdiction']))
    
    # Check 3: Date logic for IDs
    if len(customer_data['ids']) > 0:
        ids = customer_data['ids']
        for idx, id_row in ids.iterrows():
            if id_row['verification_date'] > id_row['expiry_date']:
                issues.append("ID {0}: verification_date ({1}) > expiry_date ({2})".format(
                    id_row['id_type'],
                    id_row['verification_date'].strftime('%Y-%m-%d'),
                    id_row['expiry_date'].strftime('%Y-%m-%d')))
            
            if id_row['verification_date'] > EVALUATION_DATE:
                issues.append("ID {0}: verification_date in future".format(id_row['id_type']))
    
    # Check 4: Account opened in future
    if customer_data['account_open_date'] > EVALUATION_DATE:
        issues.append("Account open date in future: {0}".format(
            customer_data['account_open_date'].strftime('%Y-%m-%d')))
    
    # Determine status
    if len(issues) == 0:
        return {
            "status": "PASS",
            "reason_code": "PASS",
            "reason": "Data quality checks passed",
            "issues": []
        }
    else:
        return {
            "status": "FLAG",
            "reason_code": "DATA_QUALITY_ISSUE",
            "reason": "{0} data quality issue(s) found".format(len(issues)),
            "issues": issues
        }

print("✓ check_data_quality function defined")
```

# Dimension 8: Regulatory Rule Application

**Purpose:** Track which regulatory rules apply and are triggered for this customer.

**Rules tracked:**
- RC-001: Enhanced Screening Frequency for High-Risk (90-day window)
- RC-002: UK ID Validity Window (180-day remaining)
- RC-003: Offshore UBO Verification (24-month + registry)
- RC-004: EU Address Evidence Recency (24-month)

**Decision Logic:**
- No rules triggered → COMPLIANT
- Any rule triggered → NON_COMPLIANT

**Output:**
- regulatory_compliance: COMPLIANT / NON_COMPLIANT
- triggered_rules: List of which rules violated (e.g., ["RC-001", "RC-003"])
- rule_details: Explanation of each trigger

```python
# Dimension 8: Regulatory Rule Application

def check_regulatory_compliance(customer_data, id_check, screening_check, ubo_check, poa_check):
    """
    Track which regulatory rules apply and are triggered.
    
    Takes results from other dimension checks as input to avoid redundant evaluation.
    
    Returns dict with regulatory compliance status and triggered rules.
    """
    
    jurisdiction = customer_data['jurisdiction']
    risk_rating = customer_data['risk_rating']
    entity_type = customer_data['entity_type']
    
    triggered_rules = []
    rule_details = []
    
    # RC-001: Enhanced Screening Frequency for High-Risk (90-day window)
    if risk_rating == "High":
        if screening_check['status'] == 'ESCALATE' and screening_check['reason_code'] == 'SCREENING_OVERDUE':
            triggered_rules.append('RC-001')
            rule_details.append('RC-001: High-risk screening overdue (90-day threshold violated)')
    
    # RC-002: UK ID Validity Window (180-day remaining)
    if jurisdiction == "UK":
        if id_check['status'] == 'ESCALATE' and 'UK rule' in id_check['reason']:
            triggered_rules.append('RC-002')
            rule_details.append('RC-002: UK ID validity window violated (<180 days remaining)')
    
    # RC-003: Offshore UBO Verification (24-month + registry)
    if entity_type == 'LEGAL_ENTITY' and jurisdiction == 'Offshore':
        if ubo_check['status'] == 'ESCALATE' and ubo_check['reason_code'] == 'UBO_STALE':
            triggered_rules.append('RC-003')
            rule_details.append('RC-003: Offshore UBO verification stale (>24 months)')
    
    # RC-004: EU Address Evidence Recency (24-month)
    if jurisdiction == "EU":
        if poa_check['status'] == 'ESCALATE':
            triggered_rules.append('RC-004')
            rule_details.append('RC-004: EU PoA evidence recency requirement violated')
    
    # Determine compliance status
    if len(triggered_rules) == 0:
        return {
            "regulatory_compliance": "COMPLIANT",
            "triggered_rules": [],
            "rule_details": []
        }
    else:
        return {
            "regulatory_compliance": "NON_COMPLIANT",
            "triggered_rules": triggered_rules,
            "rule_details": rule_details
        }

print("✓ check_regulatory_compliance function defined")
```

# Master Function: evaluate_customer_kyc()

This function performs a COMPLETE KYC evaluation for ONE customer.

**Workflow:**
1. Load all customer data (single pass)
2. Run all 8 dimension checks in sequence
3. Combine results into single output row
4. Determine overall status (ESCALATE > REMEDIATE > PASS)
5. Return comprehensive result dictionary

**Output:** One row per customer with:
- Customer info (ID, name, jurisdiction, risk_rating)
- All 8 dimension statuses + reasons
- Overall status
- Regulatory compliance status
- Data quality flags

**Performance:** Single execution per customer, no redundant lookups

```python
# CELL 12: Master Function - Comprehensive KYC Evaluation

def evaluate_customer_kyc(customer_id):
    """
    Comprehensive KYC evaluation for ONE customer across all 8 dimensions.
    
    Returns dict with complete evaluation results.
    Single pass: loads data once, runs all checks, combines output.
    """
    
    # Step 1: Load all customer data (single pass)
    customer_data = get_customer_data(customer_id)
    
    if customer_data is None:
        return {
            'customer_id': customer_id,
            'status': 'ERROR',
            'reason': 'Customer not found'
        }
    
    # Step 2: Run all 8 dimension checks
    id_check = check_id_verification(customer_data)
    poa_check = check_proof_of_address(customer_data)
    screening_check = check_screening(customer_data)
    ubo_check = check_beneficial_ownership(customer_data)
    activity_check = check_account_activity(customer_data)
    kyc_review_check = check_kyc_review(customer_data)
    data_quality_check = check_data_quality(customer_data)
    regulatory_check = check_regulatory_compliance(customer_data, id_check, screening_check, ubo_check, poa_check)
    
    # Step 3: Determine overall status (priority: ESCALATE > REMEDIATE > PASS)
    all_statuses = [
        id_check['status'],
        poa_check['status'],
        screening_check['status'],
        ubo_check['status'],
        activity_check['status'],
        kyc_review_check['status']
    ]
    
    # Remove N/A and FLAG statuses from priority logic
    priority_statuses = [s for s in all_statuses if s not in ['N/A', 'FLAG']]
    
    if 'ESCALATE' in priority_statuses:
        overall_status = 'ESCALATE'
    elif 'REMEDIATE' in priority_statuses:
        overall_status = 'REMEDIATE'
    else:
        overall_status = 'PASS'
    
    # Determine overall reason (primary escalation reason)
    overall_reason = ''
    if overall_status == 'ESCALATE':
        for check_name, check_result in [
            ('ID', id_check),
            ('PoA', poa_check),
            ('Screening', screening_check),
            ('KYC Review', kyc_review_check)
        ]:
            if check_result['status'] == 'ESCALATE':
                overall_reason = "{0}: {1}".format(check_name, check_result['reason_code'])
                break
    elif overall_status == 'REMEDIATE':
        for check_name, check_result in [
            ('ID', id_check),
            ('PoA', poa_check),
            ('Screening', screening_check),
            ('Activity', activity_check),
            ('KYC Review', kyc_review_check)
        ]:
            if check_result['status'] == 'REMEDIATE':
                overall_reason = "{0}: {1}".format(check_name, check_result['reason_code'])
                break
    else:
        overall_reason = 'All KYC checks passed'
    
    # Step 4: Combine all results into single output row
    return {
        # Customer info
        'customer_id': customer_data['customer_id'],
        'customer_name': customer_data['customer_name'],
        'entity_type': customer_data['entity_type'],
        'jurisdiction': customer_data['jurisdiction'],
        'risk_rating': customer_data['risk_rating'],
        
        # Overall status
        'overall_status': overall_status,
        'overall_reason': overall_reason,
        
        # Dimension 1: Identity Verification
        'id_status': id_check['status'],
        'id_reason_code': id_check['reason_code'],
        'id_reason': id_check['reason'],
        
        # Dimension 2: Proof of Address
        'poa_status': poa_check['status'],
        'poa_reason_code': poa_check['reason_code'],
        'poa_reason': poa_check['reason'],
        
        # Dimension 3: Screening
        'screening_status': screening_check['status'],
        'screening_reason_code': screening_check['reason_code'],
        'screening_reason': screening_check['reason'],
        
        # Dimension 4: UBO
        'ubo_status': ubo_check['status'],
        'ubo_reason_code': ubo_check['reason_code'],
        'ubo_reason': ubo_check['reason'],
        
        # Dimension 5: Activity
        'activity_status': activity_check['status'],
        'activity_reason': activity_check['reason'],
        
        # Dimension 6: KYC Review
        'kyc_review_status': kyc_review_check['status'],
        'kyc_review_reason_code': kyc_review_check['reason_code'],
        'kyc_review_reason': kyc_review_check['reason'],
        
        # Dimension 7: Data Quality
        'data_quality_status': data_quality_check['status'],
        'data_quality_issues': ', '.join(data_quality_check['issues']) if data_quality_check['issues'] else 'None',
        
        # Dimension 8: Regulatory Compliance
        'regulatory_compliance': regulatory_check['regulatory_compliance'],
        'triggered_rules': ', '.join(regulatory_check['triggered_rules']) if regulatory_check['triggered_rules'] else 'None'
    }

print("✓ evaluate_customer_kyc master function defined")
```
